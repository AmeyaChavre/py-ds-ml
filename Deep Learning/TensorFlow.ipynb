```python
import numpy as numpys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('fake_reg.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['feature1','feature2']].values
```

```python
y = df['price'].values 
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test =scaler.transform(X_test)
```

```python
X_train.min()
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
#model = Sequential([Dense(4,activation='relu'),Dense(2,activation='relu'),Dense(1)])
```

```python
model = Sequential()
```

```python
model.add(Dense(4,activation='relu'))
```

```python
model.add(Dense(2,activation='relu'))
```

```python
model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=1000)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.boxplot()
```

```python
loss_df.plot()
```

```python
loss_df.hist(color='y',edgecolor='black')
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
test_predictions =  model.predict(X_test)
```

```python
test_predictions
```

```python
test_predictions = pd.Series(test_predictions.reshape(300,))
```

```python
test_predictions
```

```python
pred_df = pd.DataFrame(y_test,columns=['Test True Y'])
```

```python
pred_df = pd.concat([pred_df,test_predictions],axis=1)
```

```python
pred_df.columns = ['Test True Y','Model Predictions']
```

```python
pred_df
```

```python
sns.scatterplot(data=pred_df,x='Test True Y',y='Model Predictions')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
df.describe()
```

```python
mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'])
```

```python
# RMS Error
mean_squared_error(pred_df['Test True Y'],pred_df['Model Predictions'])**0.5
```
