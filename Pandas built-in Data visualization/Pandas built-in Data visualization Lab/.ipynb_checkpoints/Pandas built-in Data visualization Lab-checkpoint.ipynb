
# Pandas Data Visualization 



```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

** Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remember back to your matplotlib lecture...**

```python
df3.plot.scatter(x='a',y='b',color='red',figsize=(12,3),s=50,edgecolors='black')
```

** Create a histogram of the 'a' column.**

```python
df3['a'].plot.hist(color='blue',edgecolor='black')
```

** These plots are okay, but they don't look very polished. Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it.***

```python
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(bins=30,alpha=0.5)
```

** Create a boxplot comparing the a and b columns.**

```python
df3[['a','b']].plot.box()
```

** Create a kde plot of the 'd' column **

```python
df3['d'].plot.kde()
```

** Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line)**

```python
df3['d'].plot.density(ls='--',lw=4)
```

** Create an area plot of all the columns for just the rows up to 30. (hint: use .ix).**

```python
df3.iloc[0:30].plot.area(alpha=0.4)
```

```python
f = plt.figure()
df3.iloc[0:30].plot.area(alpha=0.4,ax=f.gca())
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
