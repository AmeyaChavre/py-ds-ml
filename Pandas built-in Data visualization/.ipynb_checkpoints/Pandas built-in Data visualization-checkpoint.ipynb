# Pandas built-in Data visualization

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

### Histogram

```python
df1['A'].hist(bins=30)
```

```python
df1['A'].plot(kind='hist',bins=30)
```

```python
df1['A'].plot.hist(bins=30)
```

### Area Plot

```python
df2.plot.area(alpha=0.4)
```

### Bar Plot

```python
df2.plot.bar(stacked=True)
```

```python
df1['A'].plot.hist(bins=30)
```

### Line Plot

```python
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

### Scatter Plot

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='magma')
```

```python
df1.plot.scatter(x='A',y='B',s = df1['C'] * 50)
```

### Box Plot

```python
df2.plot.box()
```

### Hex Plot

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='coolwarm')
```

### Kernel Density Estimate Plot 

```python
# Method 1
df2['a'].plot.kde()
```

```python
# Method 2
df2['a'].plot.density()
```

```python
df2.plot.density()
```
