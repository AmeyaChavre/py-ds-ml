# Statistical Analysis using Seaborn 

## Distribution Plots

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips
```

```python
tips.head()
```

```python
# Kernel Density Estimate
sns.distplot(tips['total_bill'])
```

```python
# Removing Kernel Density Plot by setting it to "False"
# We get a histogram
sns.distplot(tips['total_bill'],kde=False,bins=40)
```

### Joint Plots

```python
# JointPlot
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
# HexPlot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
# RegressionPlot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# 2D-Plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde',)
```

### Pair Plot

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

#### RugPlot

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

## Categorical Plots

```python
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### BarPlot

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### CountPlot

```python
sns.countplot(x='sex',data=tips)
```

### BoxPlots 

```python
# Dots in the BoxPlot below represent outliers
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

### StripPlot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

### SwarmPlot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

### FactorPlot

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
```

## Matrix Plots

### Heat Maps

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
fc = flights.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
sns.heatmap(fc,annot=True,cmap='coolwarm')
```

```python
fc1 = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fc1
```

```python
sns.heatmap(fc1,cmap='magma')
```
