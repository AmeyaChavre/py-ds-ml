## Categorical Plots

```python
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### BarPlot

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### CountPlot

```python
sns.countplot(x='sex',data=tips)
```

### BoxPlots 

```python
# Dots in the BoxPlot below represent outliers
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

### StripPlot

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',dodge=True)
```

### SwarmPlot

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

### FactorPlot

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm')
```

## Matrix Plots

### Heat Maps

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
fc = flights.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
sns.heatmap(fc,annot=True,cmap='coolwarm')
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=3)
```

### Cluster Map

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

## Iris Dataset

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

### Pair Plot

```python
sns.pairplot(iris)
```

### Pair Grid

```python
g = sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### Facet Grid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

## Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

### Linear Model Plot

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],height=7,scatter_kws={"s":100},line_kws={"lw":3})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',aspect=0.7)
```
