```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[10,20,30]})
```

## Interactive Plots using Plotly 

### Line Plot

```python
# Offline Mode 
df.iplot()
```

### Scatter Plot 

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',color='blue',size=10)
```

### Bar Plot 

```python
df2.iplot(kind='bar',x='Category',y='Values',color='green')
```

```python
df.count().iplot(kind='bar',color='blue')
```

```python
df.sum().iplot(kind='bar',color='red')
```

### Box Plot

```python
df.iplot(kind='box')
```

```python
# Creating a new DataFrame

df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,40,50],'z':[100,200,300,400,500]})
```

```python
df3
```

### Surface Plot

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

### Histogram 

```python
df['A'].iplot(kind='hist',bins=50,color='blue')
```

```python
# Overlapped Histogram - each one can be turned on / off by clicking on legend

df.iplot(kind='hist')
```

### Spread Plot 

```python
import numpy as np
df[['A','B']].iplot(kind='spread')
```

### Bubble Plot

```python
df.iplot(kind='bubble',x='A',y='B',size='C',color='orange')
```

### Scatter Matrix Plot

```python
df.scatter_matrix()
```
