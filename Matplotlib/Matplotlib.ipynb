```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# <Functional Method>
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
```

```python
#Subplots
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
# <Object Oriented Method>
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Graph')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')

axes1.set_title('MAINPLOT')
axes2.set_title('SUBPLOT')
```
