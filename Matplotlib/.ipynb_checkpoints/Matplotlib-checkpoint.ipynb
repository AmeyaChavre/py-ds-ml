# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# To display plots inside Jupyter Notebook
%matplotlib inline
```

```python
# Outside Jupyter Notebook 
# Use: 
# plt.show()
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# <Functional Method>
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
```

```python
#Subplots
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
# <Object Oriented Method>
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)

axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Graph')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')

axes1.set_title('MAINPLOT')
axes2.set_title('SUBPLOT')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

#for current_axes in axes:
#    current_axes.plot(x,y)

axes[0].plot(x,y)
axes[0].plot(y,x)
axes[0].set_title('First Plot')

axes[1].plot(x,y)
axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2),dpi=100)

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,'r')
ax.plot(y,x,'g')
ax.plot(x*4,y*4,'y')
ax.plot(y*8,x*8,'b')

ax.set_title('MAIN GRAPH')

ax.set_xlabel('X-axis')

ax.set_ylabel('Y-axis')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)

axes[0].plot(x,y,'r')
axes[0].set_title('Subplot1')

axes[1].plot(x,y,'g')
axes[1].set_title('Subplot2')

plt.tight_layout()
```

```python
fig.savefig('graph.png',dpi=200)
```
